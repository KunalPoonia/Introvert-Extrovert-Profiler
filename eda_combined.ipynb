# PART B: DATA CLEANING

## Step 13: Handle Missing Values

```python
# Create a copy to preserve original data
clean_data = data.copy()
print('Created copy for cleaning!')

# Define column types
num_cols = ['Time_spent_Alone', 'Social_event_attendance', 
            'Going_outside', 'Friends_circle_size', 'Post_frequency']
cat_cols = ['Stage_fear', 'Drained_after_socializing']

# Fill numerical columns with GROUP-WISE MEDIAN
print('\nFilling numerical columns with GROUP-WISE MEDIAN:')
for col in num_cols:
    before = clean_data[col].isnull().sum()
    clean_data[col] = clean_data[col].fillna(
        clean_data.groupby('Personality')[col].transform('median')
    )
    print(f'  {col}: {before} -> 0')
```

```python
# Fill categorical columns with GROUP-WISE MODE
print('Filling categorical columns with GROUP-WISE MODE:')
for col in cat_cols:
    before = clean_data[col].isnull().sum()
    clean_data[col] = clean_data[col].fillna(
        clean_data.groupby('Personality')[col].transform(lambda x: x.mode()[0])
    )
    print(f'  {col}: {before} -> 0')

print('\nMissing values after cleaning:')
print(clean_data.isnull().sum())
```

## Step 14: Remove Duplicates

```python
before = clean_data.duplicated().sum()
clean_data = clean_data.drop_duplicates().reset_index(drop=True)
after = clean_data.duplicated().sum()

print('Duplicates before:', before)
print('Duplicates after:', after)
print('Index reset: Yes')
```

## Step 15: Drop ID Column

```python
# Drop 'id' column - not useful for analysis
print('Dropping id column...')
clean_data = clean_data.drop('id', axis=1)
print('Columns:', clean_data.columns.tolist())
```

## Step 16: Verify Data Ranges

```python
# Check data ranges for logical bounds
print('Verifying data ranges:')
for col in num_cols:
    print(f'{col}: {clean_data[col].min()} to {clean_data[col].max()}')
```

## Step 17: Check Outliers

```python
# Box plots for outlier detection
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

for i in range(len(num_cols)):
    axes[i].boxplot(clean_data[num_cols[i]])
    axes[i].set_title(num_cols[i])

axes[5].axis('off')
plt.suptitle('Checking Outliers')
plt.tight_layout()
plt.show()
```

## Step 18: Melted Boxplot - All Features by Personality

```python
# Reshape data to long format for combined visualization
# Only use numerical columns (categorical can't be mixed in boxplot)
df_long = pd.melt(
    clean_data,
    id_vars=['Personality'],
    value_vars=num_cols,  # Only numerical columns
    var_name='Feature',
    value_name='Value'
)

# Create boxplot with hue
plt.figure(figsize=(14, 7))
sns.boxplot(x='Feature', y='Value', hue='Personality', data=df_long)
plt.title('Distribution of Numerical Features by Personality')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Personality')
plt.tight_layout()
plt.show()
```

## Step 26: Prepare Features and Target

**What we need to do:**
- Separate the feature columns (X) from the target column (y)
- Features: All behavioral metrics (numerical + categorical)
- Target: `Personality` column (Extrovert/Introvert)

**Features to use:**
- `Time_spent_Alone` - Hours spent alone daily
- `Social_event_attendance` - Frequency of social events
- `Going_outside` - Frequency of going outside
- `Friends_circle_size` - Number of close friends
- `Post_frequency` - Social media posting frequency
- `Stage_fear` - Has stage fright (Yes/No)
- `Drained_after_socializing` - Feels drained after socializing (Yes/No)

## Step 27: Encode Categorical Variables

**What we need to do:**
- Convert categorical variables to numerical format for ML algorithms
- `Stage_fear`: Yes  1, No  0
- `Drained_after_socializing`: Yes  1, No  0
- `Personality` (target): Extrovert  1, Introvert  0

**Why encoding is needed:**
- Most ML algorithms require numerical input
- Label encoding works well for binary categories

## Step 28: Train-Test Split

**What we need to do:**
- Split data into training set (to train the model) and test set (to evaluate)
- Common split ratios: 80/20 or 70/30
- Use `stratify` parameter to maintain class distribution in both sets

**Why stratified split:**
- Ensures both train and test sets have similar proportion of Extroverts/Introverts
- Important when dealing with imbalanced classes

## Step 29: Feature Scaling

**What we need to do:**
- Standardize or normalize features to a common scale
- Options: StandardScaler (mean=0, std=1) or MinMaxScaler (0 to 1)

**Why scaling is important:**
- Some algorithms (SVM, KNN, Logistic Regression) are sensitive to feature scales
- Tree-based models (Decision Tree, Random Forest) don't require scaling
- Ensures all features contribute equally to the model

**Note:** Fit scaler on training data only, then transform both train and test

## Step 30: Train Classification Models

**Models to train:**

| Model | Description | Best For |
|-------|-------------|----------|
| **Logistic Regression** | Linear classifier, outputs probabilities | Baseline, interpretable |
| **Decision Tree** | Tree-based, splits on feature values | Visualization, no scaling needed |
| **Random Forest** | Ensemble of decision trees | High accuracy, handles imbalance |
| **Support Vector Machine (SVM)** | Finds optimal hyperplane | Binary classification |
| **K-Nearest Neighbors (KNN)** | Classifies based on nearest points | Small datasets |
| **XGBoost** | Gradient boosting algorithm | Best performance |

**What we'll do:**
- Train each model on the training data
- Store models for evaluation

## Step 31: Evaluate Models

**Evaluation Metrics:**

| Metric | Description | When to Use |
|--------|-------------|-------------|
| **Accuracy** | % of correct predictions | Balanced classes |
| **Precision** | TP / (TP + FP) | When false positives are costly |
| **Recall** | TP / (TP + FN) | When false negatives are costly |
| **F1-Score** | Harmonic mean of Precision & Recall | Imbalanced classes |
| **Confusion Matrix** | Shows TP, TN, FP, FN | Visual understanding |
| **ROC-AUC** | Area under ROC curve | Overall model performance |

**What we'll generate:**

- Confusion matrix visualization
- ROC curves comparison
